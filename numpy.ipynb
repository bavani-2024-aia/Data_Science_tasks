```python
import numpy as np

data_list = [10, 20, 30, 40, 50]
my_array = np.array(data_list)
print("1. Created NumPy Array:", my_array)

print("2. Data Type:", my_array.dtype)
print("   Shape of Array:", my_array.shape)
print("   Total Dimensions:", my_array.ndim)

doubled_array = my_array * 2
print("3. Array Multiplied by 2:", doubled_array)

generated_array = np.arange(1, 6)
print("4. Automatically Generated Array:", generated_array)
```

```python
import pandas as pd
import numpy as np


data = {'Name': ['Abi', 'Sam', 'Abi'], 'Age': [20, np.nan, 20]}
df = pd.DataFrame(data)
print("Raw Data:\n", df)


df = df.drop_duplicates()


df = df.fillna(0)
print("\nClean Data:\n", df)
```
