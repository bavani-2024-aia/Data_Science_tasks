```python
import pandas as pd

movies_data = {
    'movie_id': [101, 102, 103],
    'title': ['Inception', 'Avatar', 'Interstellar'],
    'year': [2010, 2009, 2014]
}
df_movies = pd.DataFrame(movies_data)

# 2. Create the 'directors' table
directors_data = {
    'movie_id': [101, 102, 104],  # Note: 104 is a different movie
    'director_name': ['Christopher Nolan', 'James Cameron', 'Steven Spielberg']
}
df_directors = pd.DataFrame(directors_data)

# 3. Perform an INNER JOIN
df_joined = pd.merge(df_movies, df_directors, on='movie_id', how='inner')

print("--- Joined Table (Only matching IDs) ---")
print(df_joined)
```
