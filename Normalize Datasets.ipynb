```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler


data = {
    'Score': [69, 79, 89, 99],
    'Salary': [30000, 45000, 70000, 110000]
}

df = pd.DataFrame(data)

scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

print(df_normalized)

```
