```python
import pandas as pd

data = {
    "Month": ["Jan", "Feb", "Mar", "Apr"],
    "Sales": [120, 150, 180, 160]
}

df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt

plt.bar(df["Month"], df["Sales"])

plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot(df["Month"], df["Sales"], marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.pie(
    df["Sales"],
    labels=df["Month"],
    autopct="%1.1f%%"
)

plt.title("Sales Distribution")

plt.show()
```

```python
import matplotlib.pyplot as plt

ages = [22, 25, 30, 35]
salary = [30000, 40000, 50000, 65000]

plt.scatter(ages, salary)

plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")

plt.show()
```

```python
import matplotlib.pyplot as plt

marks = [65, 70, 72, 75, 80, 82, 85, 90, 95]

plt.hist(marks, bins=5)

plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()
```

```python
import matplotlib.pyplot as plt

salary = [30000, 35000, 40000, 45000, 50000, 100000]

plt.boxplot(salary)

plt.title("Salary Distribution")

plt.show()
```
