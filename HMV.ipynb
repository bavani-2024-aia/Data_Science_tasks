```python
import pandas as pd
import numpy as np

data = {'Name': ['Abi', 'Sam', 'Raj'], 'Score': [90, np.nan, 80]}
df = pd.DataFrame(data)
print("Raw Data:\n", df)


df_filled = df.fillna(0)
print("\nAfter Filling with 0:\n", df_filled)


df_dropped = df.dropna()
print("\nAfter Dropping Empty Rows:\n", df_dropped)
```
