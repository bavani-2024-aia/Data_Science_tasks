```python
import pandas as pd

# Movies Table
movies = {
    "movie_id": [101, 102, 103],
    "title": ["Inception", "Avatar", "Interstellar"],
    "year": [2010, 2009, 2014]
}

df_movies = pd.DataFrame(movies)

# Directors Table
directors = {
    "movie_id": [101, 102, 104],
    "director_name": [
        "Christopher Nolan",
        "James Cameron",
        "Steven Spielberg"
    ]
}

df_directors = pd.DataFrame(directors)

# Inner Join
joined = pd.merge(
    df_movies,
    df_directors,
    on="movie_id",
    how="inner"
)

print(joined)
```
